```python
import pandas as pd

# Define the mapping from properties to (k, edge_group)
property_config = {
    'C': (50, 'doi_name'),
    'B': (5, 'doi_bioActivity'),
    'S': (50, 'doi_collectionSpecie'),
    'L': (20, 'doi_collectionSite'),
    'T': (1, 'doi_collectionType')
}

# Get the unique algorithms from the DataFrame
algorithms = df['algorithm'].unique()

# Generate the summary DataFrames
summary_dfs = {}

for algo in algorithms:
    summary = pd.DataFrame(index=property_config.keys(), columns=['1st', '2nd', '3rd', '4th'])
    
    for prop, (k_val, edge) in property_config.items():
        filtered = df[
            (df['algorithm'] == algo) &
            (df['k'] == k_val) &
            (df['edge_group'] == edge)
        ]
        # Get the maximum value for each stage
        max_values = filtered.groupby('dynamic_stage')['value'].max()
        for stage in ['1st', '2nd', '3rd', '4th']:
            summary.loc[prop, stage] = max_values.get(stage, 0.0)
    
    summary_dfs[algo] = summary.astype(float)

# Access the result for each algorithm:
# summary_dfs['deep_walk'], summary_dfs['regularization_2'], etc.
```
